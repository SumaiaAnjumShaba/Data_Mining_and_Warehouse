#**Q1**

```python
meanVector=np.mean(data3,axis=0)
# mean=data3.mean(axis=0)
print(meanVector)
#print(mean.shape)
```

#**Q2**

```python
row,col=data3.shape
n=row
Z=data3-meanVector #Center= Z=data-miu
# print(Z.shape) #(19020,10)
coVar_Matrix_inner=(1/n)*np.dot(Z.T,Z)  #1/n*(Z_T.Z)
print(coVar_Matrix_inner)
print(coVar_Matrix_inner.shape) #(10,10)

```

#**Q3**

```python
Z=data3-meanVector
n=row
coVar_Matrix_Out=np.ones([col,col])
for i in range(row):
  coVar_Matrix_Out=np.add(coVar_Matrix_Out,np.outer(Z[i].T,Z[i])) #element wise add    a = [a0, a1, ..., aM] b = [b0, b1, ..., bN]  [[a0*b0  a0*b1 ... a0*bN ]  [ ]
coVar_Matrix_Out=(1/n)*coVar_Matrix_Out
print(coVar_Matrix_Out)
```
